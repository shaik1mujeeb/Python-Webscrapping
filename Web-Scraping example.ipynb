```python
# Installing the dependencies and libraries
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import nltk
import re
```

```python
source=uReq("https://en.wikipedia.org/wiki/Global_warming").read()  #Getting the data source
```

```python
# parsing the data and creating a beautifulsoup object
soup=soup(source, "lxml")
```

```python
print(soup)
```

```python
# Fetching the data
text=""
for paragraph in soup.find_all('p'):
    text+=paragraph.text
```

```python
print(text)
```

```python
# Preprocessing the data
text=re.sub(r'\[[0-9]*\]','',text)
print(text)
```

```python
text=text.lower()
print(text)
```

```python
text=re.sub(r"\d","", text)
print(text)
```

```python
# preparing the dataset
sentences=nltk.sent_tokenize(text)
print(sentences)
```

```python
sentences=[nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
print(sentences)
```
